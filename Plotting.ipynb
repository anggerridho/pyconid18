# Plotting

```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook
init_printing()
```

The plotting module allows you to make 2-dimensional and 3-dimensional plots. Presently the plots are rendered using matplotlib as a backend. It is also possible to plot 2-dimensional plots using a TextBackend if you donâ€™t have matplotlib.

The plotting module has the following functions:

- `plot` : Plots line plots.
- `plot_parametric` : Plots 2D parametric plots.
- `plot3d` : Plots 3D plots of functions in two variables.
- `plot_implicit` : Plots 2D implicit and region plots.
- `plot3d_parametric_line` : Plots 3D line plots, defined by a parameter.
- `plot3d_parametric_surface` : Plots 3D parametric surface plots.

### `plot`

```python
# Plot
var('x')
plot(x**2)
```

```python
# Plot with range
plot(x**2,(x, 0, 10))
```

```python
# Multiple plot
p = plot(E**x, x**2, x, (x,0,4))
```

```python
# Multiple plot with color
p = plot(sin(x), cos(x), cos(x)**2 + sin(x)**2, (x, 0, pi), show=false)

p[0].line_color = 'r'
p[1].line_color = 'g'
p[2].line_color = 'b'

p.show()
```

### `plot_parametric`

```python
# Single parametric plot
plotting.plot_parametric(cos(x), sin(x), (x, -5, 5))
```

```python
# Multiple plot
plotting.plot_parametric((cos(x), sin(x)),(0.5*cos(x), 0.5*sin(x)), (x, -5, 5))
```

```python
# Multiple plot with different color
p = plotting.plot_parametric((cos(x), sin(x)),(0.5*cos(x), 0.5*sin(x)), (x, -5, 5), show=False)

p[0].line_color = 'y'
p[1].line_color = 'k'

p.show()
```

### `plot3d`

```python
var('x,y')
plotting.plot3d(x*y)
```

```python
plotting.plot3d(sin(x)*cos(y), (x, -5, 5), (y, -10, 10))
```
